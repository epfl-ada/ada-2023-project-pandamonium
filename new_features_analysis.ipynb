```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie_metadata_bechdel = pd.read_csv('movie_metadata_all_features.csv')
print(movie_metadata_bechdel.columns)
movie_metadata_bechdel=movie_metadata_bechdel.dropna(subset=["bechdel_score", "female_ratio"])
mention_group=movie_metadata_bechdel.groupby("bechdel_score")["female_ratio"].mean()

display(mention_group)

x=mention_group.index
plt.bar(x, mention_group.values, tick_label=x, label="Mean female_ratio for each bin")
plt.xlabel("Bechdel scores")
plt.ylabel("female_ratio scores")
plt.title("Bechdel scores vs actor mention scores")
plt.legend()
plt.show()
```

```python
df_bechdel = movie_metadata_bechdel.copy(deep=True)

# filter
df_bechdel_no = df_bechdel[df_bechdel['bechdel_score'] != 3]
df_bechdel_0 = df_bechdel[df_bechdel['bechdel_score'] == 0]
df_bechdel_1 = df_bechdel[df_bechdel['bechdel_score'] == 1]
df_bechdel_2 = df_bechdel[df_bechdel['bechdel_score'] == 2]
df_bechdel_3 = df_bechdel[df_bechdel['bechdel_score'] == 3]

# remove from dataframes to raw data
bechdel_no = df_bechdel_no['female_ratio'].dropna()
bechdel_0 = df_bechdel_0['female_ratio'].dropna()
bechdel_1 = df_bechdel_1['female_ratio'].dropna()
bechdel_2 = df_bechdel_2['female_ratio'].dropna()
bechdel_3 = df_bechdel_3['female_ratio'].dropna()

# Create a figure and axes
fig, ax = plt.subplots(1,2, figsize=(16,8))

# Create two boxplots in one figure
ax[0].boxplot([bechdel_0, bechdel_1, bechdel_2, bechdel_3], labels=['0', '1', '2', '3'], showfliers=False, showmeans=True)

# Set labels and title
ax[0].set_xlabel('Bechdel Score')
ax[0].set_ylabel('Female Cast Ratio')
ax[0].set_title('Bechdel Test vs Female Cast Ratio')

ax[1].boxplot([bechdel_no, bechdel_3], labels=['Fail', 'Pass'], showfliers=False, showmeans=True)

# Set labels and title
ax[1].set_xlabel('Bechdel Score')
ax[1].set_ylabel('Female Cast Ratio')
ax[1].set_title('Bechdel Test vs Female Cast Ratio (Binary)')

print('Number of Fail: {} \nNumber of Pass: {}'.format(len(bechdel_no), len(bechdel_3)))
# Show the plot
plt.show()
```
