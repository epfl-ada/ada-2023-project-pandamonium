```python
# Group by 1. Wikipedia movie ID and 6. Actor gender, then count the occurrences of each gender
gender_counts = character_metadata_bechdel.groupby(['1. Wikipedia movie ID', '6. Actor gender']).size().unstack(fill_value=0)

# Calculate the ratio of female actors to total actors for each movie
gender_counts['female_ratio'] = gender_counts['F'] / (gender_counts['M'] + gender_counts['F'])
#display(gender_counts)

# merge the gender_counts and character_metadata_bechdel
character_metadata_bechdel_ratio = gender_counts.merge(character_metadata_bechdel, \
                                                                    left_index=True, right_on='1. Wikipedia movie ID')
# print the merged DataFrame
print(character_metadata_bechdel_ratio)

# # optional:
# character_metadata_bechdel_ratio = character_metadata_bechdel_ratio.drop(columns=['F', 'M'])

unique_count_wiki_ratio = character_metadata_bechdel_ratio['1. Wikipedia movie ID'].nunique()
print("Number of unique Wikipedia movie ID values:", unique_count_wiki_ratio)
```
